```python
import tkinter as tk
from tkinter import ttk

# Function to update sign colors based on selection
def update_signal(event=None):
    signal = signal_var.get()
    if signal == "Stop":
        label_display.config(text="STOP", bg="red", fg="white")
    elif signal == "Wait":
        label_display.config(text="WAIT", bg="yellow", fg="black")
    elif signal == "Go":
        label_display.config(text="GO", bg="green", fg="white")
    else:
        label_display.config(text="", bg="white", fg="black")

# Create main window
root = tk.Tk()
root.title("Traffic Signal")
root.geometry("300x200")

# Variable to store selected signal
signal_var = tk.StringVar()

# Label: Choose Signal
label_select = tk.Label(root, text="Select Signal:", font=("Arial", 12))
label_select.pack(pady=10)

# Dropdown menu for selecting signal
signal_menu = ttk.Combobox(root, textvariable=signal_var, state="readonly")
signal_menu["values"] = ("Stop", "Wait", "Go")
signal_menu.pack()
signal_menu.bind("<<ComboboxSelected>>", update_signal)

# Display label for the signal
label_display = tk.Label(root, text="", font=("Arial", 20, "bold"), width=10, height=3)
label_display.pack(pady=20)

root.mainloop()
```
