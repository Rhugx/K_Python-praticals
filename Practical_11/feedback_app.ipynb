```python
from flask import Flask, render_template_string, request

app = Flask(__name__)

# HTML form page
feedback_form = """
<!DOCTYPE html>
<html>
<head>
    <title>Feedback Page</title>
</head>
<body style="font-family: Arial; background-color: #f2f2f2; padding: 20px;">
    <h2>Feedback Form</h2>
    <form method="POST" action="/submit">
        <label>Your Name:</label><br>
        <input type="text" name="name" required><br><br>

        <label>Your Email:</label><br>
        <input type="email" name="email" required><br><br>

        <label>Your Feedback:</label><br>
        <textarea name="comments" rows="5" cols="30" required></textarea><br><br>

        <input type="submit" value="Submit Feedback">
    </form>
</body>
</html>
"""

# Thank-you page
thank_you_page = """
<!DOCTYPE html>
<html>
<head>
    <title>Thank You</title>
</head>
<body style="font-family: Arial; background-color: #e6ffe6; padding: 20px;">
    <h2>Thank You, {{name}}!</h2>
    <p>We appreciate your feedback:</p>
    <blockquote>{{comments}}</blockquote>
    <p>We will contact you at: <b>{{email}}</b></p>
</body>
</html>
"""

@app.route('/')
def home():
    return render_template_string(feedback_form)

@app.route('/submit', methods=['POST'])
def submit_feedback():
    name = request.form['name']
    email = request.form['email']
    comments = request.form['comments']
    return render_template_string(thank_you_page, name=name, email=email, comments=comments)

if __name__ == "__main__":
    app.run(host="127.0.0.1", port=5000, debug=False, use_reloader=False)
```
