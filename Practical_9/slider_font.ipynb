```python
import tkinter as tk
from tkinter import font

# Function to update font size
def update_font_size(value):
    size = int(value)
    label.config(font=("Arial", size))

# Create main application window
root = tk.Tk()
root.title("Font Size Controller")
root.geometry("400x250")

# Create a Label widget
label = tk.Label(root, text="Adjust my size!", font=("Arial", 20))
label.pack(pady=40)

# Create a Slider (Scale) widget
slider = tk.Scale(root,
                  from_=8,         # Minimum font size
                  to=50,           # Maximum font size
                  orient="horizontal",  # Horizontal slider
                  command=update_font_size)
slider.set(20)  # Default font size
slider.pack()

# Run the Tkinter event loop
root.mainloop()
```
