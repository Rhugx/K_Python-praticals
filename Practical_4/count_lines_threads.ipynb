```python
import threading

# Function to count lines in a file
def count_lines(filename):
    try:
        with open(filename, 'r') as file:
            line_count = sum(1 for _ in file)
        print(f"[{threading.current_thread().name}] '{filename}' has {line_count} lines.")
    except FileNotFoundError:
        print(f"[{threading.current_thread().name}] File '{filename}' not found.")

# Create sample test files
with open("file1.txt", "w") as f1:
    f1.write("""Hello, this is file 1.
It contains a few lines.
Python threading example.
Count lines quickly.
That's it for file 1.""")

with open("file2.txt", "w") as f2:
    f2.write("""This is file 2.
It also contains some lines.
Thread 2 will read this.
Multithreading makes it faster.
End of file 2.""")

# File names
file1 = "file1.txt"
file2 = "file2.txt"

# Create threads
thread1 = threading.Thread(target=count_lines, args=(file1,), name="Thread-File1")
thread2 = threading.Thread(target=count_lines, args=(file2,), name="Thread-File2")

# Start threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("✅ Line counting completed.")
```
