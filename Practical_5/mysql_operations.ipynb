```python
import mysql.connector

# Database connection details
HOST = "127.0.0.1"
USER = "root"
PASSWORD = ""  # XAMPP default: no password
DATABASE = "school"

# 1. Connect to MySQL server
conn = mysql.connector.connect(
    host=HOST,
    user=USER,
    password=PASSWORD
)
cursor = conn.cursor()
print("‚úÖ Connected to MySQL server.")

# 2. Create database if not exists
cursor.execute(f"CREATE DATABASE IF NOT EXISTS {DATABASE}")
print(f"‚úÖ Database '{DATABASE}' is ready.")

# Switch to the database
cursor.execute(f"USE {DATABASE}")

# 3. Drop table if exists
cursor.execute("DROP TABLE IF EXISTS Students")
print("üóëÔ∏è Dropped existing 'Students' table if it existed.")

# 4. Create table
cursor.execute("""
    CREATE TABLE Students (
        id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(50),
        age INT,
        grade VARCHAR(10)
    )
""")
print("‚úÖ Table 'Students' created fresh.")

# 5. Insert data
insert_query = "INSERT INTO Students (name, age, grade) VALUES (%s, %s, %s)"
students = [
    ("Alice", 20, "A"),
    ("Bob", 21, "B"),
    ("Charlie", 19, "A")
]
cursor.executemany(insert_query, students)
conn.commit()
print(f"‚úÖ Inserted {cursor.rowcount} records into Students.")

# 6. Read data
cursor.execute("SELECT * FROM Students")
rows = cursor.fetchall()
print("\nüìÑ Student Records:")
for row in rows:
    print(row)

# 7. Update data
cursor.execute("UPDATE Students SET grade = %s WHERE name = %s", ("A+", "Bob"))
conn.commit()
print(f"‚úÖ Updated grade for Bob.")

# 8. Delete data
cursor.execute("DELETE FROM Students WHERE name = %s", ("Charlie",))
conn.commit()
print(f"‚úÖ Deleted record for Charlie.")

# 9. Display updated table
cursor.execute("SELECT * FROM Students")
rows = cursor.fetchall()
print("\nüìÑ Updated Student Records:")
for row in rows:
    print(row)

# 10. Close connection
cursor.close()
conn.close()
print("üîí MySQL connection closed.")
```
