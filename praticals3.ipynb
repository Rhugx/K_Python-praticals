```python
import threading
import math
import time

# Thread function to calculate factorial
def calc_factorial(num):
    print(f"[Factorial Thread] Calculating factorial of {num}...")
    time.sleep(1)  # Simulate delay
    result = math.factorial(num)
    print(f"[Factorial Thread] Factorial of {num} is {result}")

# Thread function to calculate square
def calc_square(num):
    print(f"[Square Thread] Calculating square of {num}...")
    time.sleep(1)
    result = num * num
    print(f"[Square Thread] Square of {num} is {result}")

# Thread function to calculate summation of list
def calc_sum(numbers):
    print(f"[Summation Thread] Calculating sum of {numbers}...")
    time.sleep(1)
    result = sum(numbers)
    print(f"[Summation Thread] Sum of list is {result}")

# Main program
if __name__ == "__main__":
    # Sample data
    number_for_factorial = 5
    number_for_square = 7
    list_for_sum = [1, 2, 3, 4, 5]

    # Create threads
    t1 = threading.Thread(target=calc_factorial, args=(number_for_factorial,))
    t2 = threading.Thread(target=calc_square, args=(number_for_square,))
    t3 = threading.Thread(target=calc_sum, args=(list_for_sum,))

    # Start threads
    t1.start()
    t2.start()
    t3.start()

    # Wait for all threads to finish
    t1.join()
    t2.join()
    t3.join()

    print("[Main Thread] All calculations completed.")
```
