```python
import re

# Sample text lines
lines = [
    "From john.doe@example.com Sat Jan  5 09:14:16 2008",
    "From jane_smith@domain.org Tue Feb 12 14:25:30 2010"
]

# Regex patterns
pattern_timestamp = r'\b(?:Mon|Tue|Wed|Thu|Fri|Sat|Sun)\s+[A-Z][a-z]{2}\s+\d{1,2}\s+\d{2}:\d{2}:\d{2}\s+\d{4}'
pattern_email = r'\b[\w\.-]+@[\w\.-]+\.\w+\b'
pattern_month = r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b'
pattern_year = r'\b\d{4}\b'
pattern_time = r'\b\d{2}:\d{2}:\d{2}\b'

# Processing each line
for line in lines:
    print(f"Line: {line}")
    
    # a) Complete timestamps
    timestamps = re.findall(pattern_timestamp, line)
    print("  a) Timestamps:", timestamps)
    
    # b) Complete e-mail addresses
    emails = re.findall(pattern_email, line)
    print("  b) E-mails:", emails)
    
    # c) Only months
    months = re.findall(pattern_month, line)
    print("  c) Months:", months)
    
    # d) Only years
    years = re.findall(pattern_year, line)
    print("  d) Years:", years)
    
    # e) Only time
    times = re.findall(pattern_time, line)
    print("  e) Time:", times)
    print("-" * 50)
```
